Q1

```python
import numpy as np
import pandas as pd

rows = 20_000        # total rows
cols = 10_000        # total columns
chunk_size = 500     # rows per chunk (safe for memory)

npy_file = "big_data.npy"
csv_sample_file = "sample_data.csv"

data = np.memmap(
    npy_file,
    dtype=np.float32,
    mode="w+",
    shape=(rows, cols)
)

for start in range(0, rows, chunk_size):
    end = min(start + chunk_size, rows)


    block = np.zeros((end - start, cols), dtype=np.float32)


    block[:, :3334] = np.random.uniform(
        0, 1, size=(end - start, 3334)
    )


    block[:, 3334:6667] = np.random.normal(
        0, 1, size=(end - start, 3333)
    )


    block[:, 6667:] = np.random.randint(
        0, 100, size=(end - start, cols - 6667)
    )


    data[start:end] = block

    print(f"Saved rows {start} to {end}")


data.flush()
print("Finished creating big_data.npy")


sample_rows = 100
sample_cols = 50

sample = data[:sample_rows, :sample_cols]

df = pd.DataFrame(sample)
df.to_csv(csv_sample_file, index=False)

print(f"CSV sample saved as {csv_sample_file}")
```

```python
import numpy as np
from collections import defaultdict

arr = np.array([1, 2, 2, 3, 1, 4, 3, 2])

groups = defaultdict(list)

for val in arr:
    groups[val].append(val)


grouped_matrix = np.array(list(groups.values()), dtype=object)

print("Grouped Matrix:")
print(grouped_matrix)
```
